### <span style="color:navy"> Step III: Complete the Function </span>
After you get the color name from above as **R**, **G** or **B**, write your script in the cell below under the function named as <span style="color:navy"> **def** </span> <span style="color:green"> **conversion():** </span>, according to the instructions

````python
def conversion():
    cap = cv2.VideoCapture('videorec.mp4')
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))   # float `width`
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    print("width=", width, "height=", height)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    #fps = int(cap.get(cv2.CAP_PROP_FPS))
    out = cv2.VideoWriter('output.mp4',fourcc, 20.0, (width,height))
    
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret==0:
            break
            
        # YOUR CODE HERE
        green_channel = np.zeros_like(frame)
        green_channel[:,:,1] = frame[:,:,1]
        
        #raise NotImplementedError()
        
        ### End Code here
        #
        out.write(green_channel)
    print("output.mp4 completed")
    cap.release()
    out.release()
    return
````
