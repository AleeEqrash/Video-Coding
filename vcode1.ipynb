```python
import numpy as np
import matplotlib.pyplot as plt

fs = 44100
fft_fs = fs/2
f = 1000
amp = 1
duration = 1
x = np.arange(0,duration,1/fs)
y = amp*np.sin(2*np.pi*f* x)
plt.plot(y[1:100]) #plotting first 100 samples
plt.show()

#fourier
y_fft = np.fft.fft(y)
fft_len = len(y_fft)//2
plt.plot(-y_fft[:fft_len]) #plotting fourier of fs/2
plt.show()

```

```python
Y = np.fft.fft(y)
freq = np.fft.fftfreq(len(y), 1/fs)

# plot single-sided spectrum
plt.plot(freq[:len(freq)//2])
#plt.xlim([0, fs/2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Single-Sided Fourier Spectrum of a Sine Wave')
plt.show()
```
